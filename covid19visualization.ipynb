```python
def nice_axes(ax):
    ax.set_facecolor('.8')
    ax.tick_params(labelsize=8, length=0)
    ax.grid(True, axis='x', color='white')
    ax.set_axisbelow(True)
    [spine.set_visible(False) for spine in ax.spines.values()]
    
nice_axes(ax)
fig
```

```python
grouped = dfs.groupby(['location','date'])
covid_confirmed = grouped.sum().reset_index().sort_values(['date'],ascending=False)
covid_confirmed.head()
```

```python
dfp = (covid_confirmed[covid_confirmed['date'].eq("2020-06-01")].sort_values(by="total_cases",ascending=False).head(10))

#plotting the initial horaizontal barchart 
fig, ax = plt.subplots(figsize=(15, 8))
ax.barh(dfp['location'], dfp['total_cases'])
plt.xlabel('Number of Confirmend Cases')
plt.ylabel('Country')
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
# pass colors values to `color=`
ax.barh(dfp['location'], dfp['total_cases'], color=[colors[x] for x in dfp['location']])
# iterate over the values to plot labels and values (Tokyo, Asia, 38194.2)
for i, (value, name) in enumerate(zip(dfp['total_cases'], dfp['location'])):
    ax.text(value, i,     name,            ha='right')  # Tokyo: name
    ax.text(value, i-0.25,     value,           ha='right')   # 38194.2: value
# Add year right middle portion of canvas
ax.text(1, 0.4, "2020-06-01", transform=ax.transAxes, size=46, ha='right')
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
def draw_barchart(date):
    case="total_cases"
    dfp= (covid_confirmed[covid_confirmed['date'].eq(date)].sort_values(by=case,ascending=False).head(10))[::-1]
    ax.clear()
    ax.barh(dfp['location'], dfp['total_cases'], color=["#980505","#CD1212","#D84E4E","#CB6262","#D39B5F","#F7EC10","#D0F710","#9CF710","#B4D67F","#969C8E"][::-1])
    dx = dfp['total_cases'].max() / 200
    for i, (value, name) in enumerate(zip(dfp['total_cases'], dfp['location'])):
        ax.text(value-dx, i,     name,           size=14, weight=600, ha='right', va='bottom')
        ax.text(value-dx, i-.25, value, size=10, color='#17202A', ha='right', va='baseline')
        #ax.text(value+dx, i,     f'{value:,.0f}',  size=14, ha='left',  va='center')
    # ... polished styles
    ax.text(1, 0.4, date, transform=ax.transAxes, color='#777777', size=30, ha='right', weight=800)
    ax.text(0, 1.06, 'Number of Cases', transform=ax.transAxes, size=12, color='#777777')
    ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
    ax.xaxis.set_ticks_position('top')
    ax.tick_params(axis='x', colors='#777777', labelsize=12)
    ax.set_yticks([])
    ax.margins(0, 0.01)
    ax.grid(which='major', axis='x', linestyle='-')
    ax.set_axisbelow(True)
    ax.text(0, 1.12, 'The number of '+str(case)+' Cases upto 25 April 2020',
            transform=ax.transAxes, size=24, weight=600, ha='left')
    ax.text(1, 0, 'by @Zubair; credit @jburnmurdoch', transform=ax.transAxes, ha='right',
            color='#777777', bbox=dict(facecolor='white', alpha=0.8, edgecolor='white'))
    plt.box(False)
draw_barchart('2020-06-01')
```
